```python
import cv2
import numpy as np 
import matplotlib.pyplot as plt
```

```python
img = cv2.imread('img2.jpg')
img = cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
plt.imshow(img)
```

```python
gray = cv2.cvtColor(img,cv2.COLOR_RGB2GRAY)
plt.imshow(gray,cmap='gray')
```

```python
blr = cv2.medianBlur(gray,11)
plt.imshow(blr,cmap='gray')
```

```python
_,th = cv2.threshold(blr,180,255,cv2.THRESH_BINARY)
plt.imshow(th,cmap='gray')
```

```python
dilate = cv2.dilate(th,(11,11),iterations=7)
plt.imshow(dilate,cmap='gray')
```

```python
erode = cv2.erode(dilate,(11,11),iterations=7)
plt.imshow(erode,cmap='gray')
```

```python
res = img.copy()
defective = False
contour,heirarchy = cv2.findContours(erode,cv2.RETR_CCOMP,cv2.CHAIN_APPROX_SIMPLE)
for i in range(len(contour)):
    if (erode==0).sum() > 0:
        defective = True
        if heirarchy[0][i][3] != 0:
            cv2.drawContours(res,contour,i,(210,20,30),1)
plt.imshow(res)
```

```python
fig = plt.figure(figsize=(10,8))
if(defective):
    fig.suptitle("OUTPUT: Defective Fabric",fontsize=30,color='red')
else:
    fig.suptitle("OUTPUT: Good Fabric",fontsize=30,color='green')

fig.add_subplot(331)
plt.imshow(img)
plt.title("Original Image")
plt.xticks([])
plt.yticks([])

fig.add_subplot(332)
plt.imshow(gray,cmap='gray')
plt.title("Gray Image")
plt.xticks([])
plt.yticks([])

fig.add_subplot(333)
plt.imshow(blr,cmap='gray')
plt.title("Blured Image")
plt.xticks([])
plt.yticks([])

fig.add_subplot(334)
plt.imshow(th,cmap='gray')
plt.title("Thresholded Image")
plt.xticks([])
plt.yticks([])

fig.add_subplot(335)
plt.imshow(dilate,cmap='gray')
plt.title("Dilation Image")
plt.xticks([])
plt.yticks([])

fig.add_subplot(336)
plt.imshow(erode,cmap='gray')
plt.title("Erosion Image")
plt.xticks([])
plt.yticks([])

fig.add_subplot(337)
plt.imshow(res,cmap='gray')
plt.title("Output Image")
plt.xticks([])
plt.yticks([])

fig.tight_layout()
```
